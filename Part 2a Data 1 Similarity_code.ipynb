```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('similarity_scores.csv')
```

```python
df.info()
```

```python
df = df[['recipe_a', 'recipe_b', 'score']]
```

```python
df.head()
```

```python
# Duplicate tables and apend to create a data set for matrix
data1 = df[['recipe_a', 'recipe_b', 'score']]
data1.rename(columns={'recipe_a': 'recipe_bb'}, inplace=True)
data1.rename(columns={'recipe_b': 'recipe_a'}, inplace=True)
data1.rename(columns={'recipe_bb': 'recipe_b'}, inplace=True)
data3 = data1.append(df)
data3.info()
```

```python
table1 = pd.pivot_table(data3, index = ['recipe_a'], columns = ['recipe_b'],
                       values = ['score'], aggfunc ={'score':np.mean}).rename(columns ={'score':'recipe_id'} )
table1
```
